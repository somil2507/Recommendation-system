## constructing Pivot Table

```python
user_bk = bk.pivot_table(index = ['User.ID'],
                                 columns = ['Book.Title'],
                                 values = ['Book.Rating']).reset_index(drop=True)
```

```python
user_bk
```

```python
user_bk.index = bk["User.ID"].unique()
user_bk
```

## Filling 0 in place of NaN

```python
user_bk.fillna(0, inplace=True)
```

```python
user_bk
```

## Applying cousine function

```python
user_sim = 1 - pairwise_distances( user_bk.values,metric='cosine')
```

```python
user_sim
```

## Store the results in a dataframe

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

## Set the index and column names to user ids 

```python
user_sim_df.index = bk["User.ID"].unique()
user_sim_df.columns = bk["User.ID"].unique()
```
